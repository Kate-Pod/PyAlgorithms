### Python Algoritms

```python
# Imagine I came up with this function
def sum1(n):
    '''
    Take an input of n and return the sum of the numbers from 0 to n
    '''
    final_sum = 0
    for x in range(n+1):
        final_sum += x
    return final_sum
```

```python
sum1(5)
```

```python
# And then there is this function
def sum2(n):
    '''
    Take an input of n and return the sum of the numbers from 0 to n
    '''
    return(n*(n+1))/2
```

```python
sum2(5)
```

Function sum1 uses a for loop to iteratively add across our range +1

Function sum2 makes use of a formula to solve a problem

### Comparison

**Big-O** notation describes how quickly runtime will grow relative to the input as the input get arbitrarily large.

```python
def Bigo(n):
    return 45*n**3 + 20*n**2 + 19
Bigo(1)
```

```python
Bigo(2)
#19 will not be a scaling or limiting factor
```

```python
Bigo(10)
# The '20n**2' is 2000 and '45n**3' is 45000 = O(n**3)
```

So we want to choose abgorithms that stay away from any exponential, quadratic or cubic behaviour.
